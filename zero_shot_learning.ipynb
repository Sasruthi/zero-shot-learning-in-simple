# Importing  pipeline .

```python
from transformers import pipeline
```

# Initializing  pipeline

The imported pipeline is initialized to a variable. The result is stored in a  variable called variable_pipeline.




*   Here we are using the facebook/bart-large-mnli model. 
*   **BERT - Bidirectional Encoder Representations** from Transformers




BERT is used because it tends to generate contextualized embeddings. It looks at the statement’s context and generates a meaningful vector representation for a given word.


*   BERT Large is a huge model using 24 encoder layers.
*   BERT large is used for this categorization of text to their domain



```python
variable_pipeline = pipeline ("zero-shot-classification", model = "facebook/bart-large-mnli")
```

```python
input_sequence = "We are traveling to Himalayas"
label_candidate = ['travel', 'cooking', 'entertainment', 'dancing', 'technology']
variable_pipeline(input_sequence, label_candidate)
```

From the results, the model has correctly classified the sentence as  travel category with 96% accuracy.
